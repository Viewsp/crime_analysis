```python
#Import proper libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import folium 

%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
crimeLA = pd.read_csv('CrimeLA.csv')
crimeLA.head()
```

```python
#Remove columns with NaN values that may not be needed for our analysis since we have a good amount of 
#columns to conduct our analysis
crimeLA.drop(columns=['Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4','Cross Street','Weapon Used Cd'], inplace=True)
crimeLA.head()
```

```python
#Remove columns with NaN values that may not be needed for our analysis since we have a good amount of 
#columns to conduct our analysis
crimeLA.drop(columns=['Premis Cd'], inplace=True)
```

```python
#Remove columns with NaN values that may not be needed for our analysis since we have a good amount of 
#columns to conduct our analysis
crimeLA.drop(columns=['Mocodes'], inplace=True)
```

```python
#Convert date columns to dateTime
crimeLA['Date Rptd'] = pd.to_datetime(crimeLA['Date Rptd'])
```

```python
crimeLA['DATE OCC'] = pd.to_datetime(crimeLA['DATE OCC'])
```

```python
crimeLA.dtypes
```

```python
crimeLA.describe()
```

```python
crimeLA.describe(include = ['O'])
```

```python
#Create a dt.month column which equates January=1 & Decemeber=12
crimeLA['month'] = crimeLA['DATE OCC'].dt.month
```

```python
#Create a dt.year column
crimeLA['year'] = crimeLA['DATE OCC'].dt.year
```

```python
#Since we have a large dataset, a total of 440716 rows. We will filter for only the year 2022 and explore
crime_new = crimeLA['year'] > 2021
crime_year = crimeLA[crime_new]
crime_year.head()
```

```python
len(crime_year)
```

```python
#Mean of crimes per location within Los Angeles
crime_year.groupby(['AREA NAME']).mean()
```

```python
#Count of crimes per location within Los Angeles
crime_year.groupby(['AREA NAME']).count()
```

**We will begin by filtering the crimes per AREA NAME. Based on our description we have a total of 21 unique AREA NAMES(locations) within Los Angeles. Filter for crimes only within West LA**

```python
crime_west = crime_year['AREA NAME'] == 'West LA'
crime_westLA = crime_year[crime_west]
crime_westLA.head()
```

```python
#We have a total of 1624 rows for crimes in West LA
len(crime_westLA)
```

```python
#Filter for only crimes for the month on Jan
crime_westL_Jan = crime_westLA['month'] == 1
crime_westL_Jan = crime_westLA[crime_westL_Jan]
len(crime_westL_Jan)
```

```python
crime_westL_Jan.reset_index(inplace = True)
crime_westL_Jan
```

**Create Folium map with crimes in West LA**

```python
LA_map = folium.Map([34.0522, -118.2437])
LA_map
```

```python
for i in range(len(crime_westL_Jan)):
    row = crime_westL_Jan.loc[i]
    folium.Marker([row["LAT"], row["LON"]],
                  tooltip = row["LOCATION"],
                  popup = row['Crm Cd Desc']
             ).add_to(LA_map)
LA_map
```
